### Metric utils

Accuracy

```python
import torch
```

```python
def accuracy(predicted :torch.tensor, gt:torch.tensor):
    
    correct_preds = sum(predicted == gt)
    toal_preds = len(predicted)
    
    return (correct_preds/toal_preds)*100


def recall (predicted :torch.tensor, gt:torch.tensor, positive_class: int = 1):
    

    TP = sum((predicted==positive_class) & (gt==positive_class))
    FN = sum((predicted!=positive_class) & (gt==positive_class))
    
    eps = 10e-10
    return (TP/ (TP + FN + eps)) * 100

def precision(predicted :torch.tensor, gt:torch.tensor, positive_class: int = 1):
    
    TP = sum((predicted==positive_class) & (gt==positive_class))
    FP = sum((predicted==positive_class) & (gt!=positive_class))
    
    eps = 10e-10
    return (TP/ (TP + FP + eps)) * 100



def f1score(predicted :torch.tensor, gt:torch.tensor, positive_class: int = 1):
    p = precision(predicted, gt, positive_class)
    r = recall(predicted, gt, positive_class)
    
    return 2*(p*r/(p+r))  

```

```python
predicted = torch.tensor([0,0,0,1,1,1])
gt =        torch.tensor([0,1,0,1,0,1])

accuracy(predicted, gt), recall(predicted, gt), precision(predicted, gt), f1score(predicted, gt)
```
